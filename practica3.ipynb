```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
# Set the path to the file you'd like to load
path = kagglehub.dataset_download("tannubarot/cybersecurity-attack-and-defence-dataset")

print("Path to dataset files:", path)
```

```python
## importar datos desde el csv a un dataframe
df = pd.read_csv("Attack_Dataset.csv")
# Mostrar las primeras filas del dataframe
df.head()
```

```python
# Muestra numero total de filas(registros) y columnas.
# Muestra nombre de cada columna.
# Cantidad de valores no nulos en cada columna.
# Tipo de dato de cada columna.
# Tamano de archivo en memoria.
df.info()
```

```python
# Elimino la columna 15
df = df.drop(columns=['Unnamed: 15'])

# Eliminar las filas con los valores vacios
df = df.dropna()
# Elimina solamente las filas que tengan todos los valores nulos en sus columnas.
# df = df.dropna(how='all')
```

```python
df.info()
```

```python
# En la fila 'Tools Used' dentro de 'df' procesa cada celda individualmente con funcion lambda, y si x es una cadena de texto (str), la función la divide en una lista separando por las comas, sino deja el valor tal cual.
df['Tools Used'] = df['Tools Used'].apply(lambda x: x.split(',') if isinstance(x, str) else x)
```

```python
df.head()
```

```python
#Tipos de ataques mas comunes
attack_counts = df['Attack Type'].value_counts()
print(attack_counts)
```

```python
# Top 10 de ataques mas comunes
import seaborn as sns
import matplotlib.pyplot as plt

top10_attacks = attack_counts.head(10)
sns.barplot(x=top10_attacks.index, y=top10_attacks.values, palette='magma')
plt.xticks(rotation=90)
plt.title("Top 10 de tipos de ataque mas comunes")
plt.figure(figsize=(10, 6))
plt.show()
```

Si observamos los tipos de ataques mas comunes, el quinto ataque mas comun es "Malicious Libraries" y el sexto ataque mas comun es "Malicious Library" lo que se puede entender como un solo dato de tipos de ataques mas comunes. Tomarlos por separado podria llevar a errores de analisis, conteos incorrectos, y graficos duplicados como en el paso anterior.
