<h4>Análisis de Categorías con su impacto</h4>

```python
print(df['Impact'])
print(df['Category'])
```

```python
#expandimos los datos solo para el caso de impact
df_exploded = df.explode('Impact')
# se hace un tratamiento al texto para eliminar espacios en blanco al inicio y final de cada valor en la columna 'Impact'
df_exploded['Impact'] = df_exploded['Impact'].str.strip()
# también se deja todo en minúsculas
df_exploded['Impact'] = df_exploded['Impact'].str.lower()
```

```python
print(df_exploded['Impact'].value_counts())
print(df_exploded['Category'].value_counts())
```

Se observa que hay demasiados valores unicos ya sea por errores de tipeado del dataset o falta de comas u ortografía, entonces se ha decidido eliminar estos valores unicos para continuar el análisis y dejar solo los que tengan 2 o más coincidencias

```python
# Se eliminan las filas que tienen valores unicos en la columna 'Impact'
impact_counts = df_exploded['Impact'].value_counts()
impact_counts = impact_counts[impact_counts > 1]
df_exploded = df_exploded[df_exploded['Impact'].isin(impact_counts.index)]
df_exploded['Impact'].value_counts()
```

Al haber demasiados valores únicos en los impactos se decidió hacer un clusterizado en machine learning para que los datos gráficos sean mejor entendibles y se pueda sacar mejores conclusiones

```python
#Los impactos al ser demasiados, se decide hacer clusterización para agruparlos
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
# Convertir los textos en vectores TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_exploded['Impact'])
# Aplicar KMeans para agrupar los textos
num_clusters = 15  
kmeans = KMeans(n_clusters=num_clusters, random_state=100)
df_exploded['Impact_Cluster'] = kmeans.fit_predict(X)
# Mostrar los textos agrupados por cluster
for cluster in range(num_clusters):
    print(f"Cluster {cluster}:")
    print(df_exploded[df_exploded['Impact_Cluster'] == cluster]['Impact'].unique())
    
```

```python
# Obtener un impacto representativo para cada cluster
#cluster_labels = df_exploded.groupby('Impact_Cluster')['Impact'].agg(lambda x: x.value_counts().index[0])
#cluster_labels
cluster_names = {
    0: "Execution of Code",
    1: "System Instability & Information Leak",
    2: "Full System Compromise", 
    3: "Unauthorized Access",
    4: "Data Breach & Privacy Violation",
    5: "Session & Resource Hijacking",
    6: "Credential Theft",
    7: "Full-Scale Compromise",
    8: "Account & Control Takeover",
    9: "User Impact & Deception",
    10: "Denial-of-Service",
    11: "Security Bypass",
    12: "Malware Deployment",
    13: "Data Exfiltration & Manipulation",
    14: "Phishing Operations"
}
cluster_impact_labels = pd.Series(cluster_names)
cluster_impact_labels.index.name = 'Impact_Cluster'
cluster_impact_labels.name = 'Impact_Cluster_Name'
```

```python
# Se hace el mismo proceso de clusterizado para la columna 'Category'
df_exploded = df_exploded.explode('Category')
df_exploded['Category'] = df_exploded['Category'].str.strip()
df_exploded['Category'] = df_exploded['Category'].str.lower()
#elimino caracteres no alfanuméricos
df_exploded['Category'] = df_exploded['Category'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
print(df_exploded['Category'].value_counts())
```

```python
# Convertir los textos en vectores TF-IDF
vectorizer = TfidfVectorizer()
X_cat = vectorizer.fit_transform(df_exploded['Category'])
# Aplicar KMeans para agrupar los textos
num_clusters_cat = 25
kmeans_cat = KMeans(n_clusters=num_clusters_cat, random_state=60)
df_exploded['Category_Cluster'] = kmeans_cat.fit_predict(X_cat)
# Mostrar los textos agrupados por cluster
for cluster in range(num_clusters_cat):
    print(f"Cluster {cluster}:")
    print(df_exploded[df_exploded['Category_Cluster'] == cluster]['Category'].unique())
```

```python
cluster_category_names = pd.Series({
    0: "Web Application Security",
    1: "Physical & Hardware Attacks",
    2: "Blockchain & Web3 Security",
    3: "Browser Security & Client-Side Attacks",
    4: "Insider Threats",
    5: "Operating System Exploits",
    6: "Cryptography & Quantum Threats",
    7: "AI/ML Security",
    8: "AI Data Leakage & Privacy Risks",
    9: "Email & Messaging Protocol Exploits",
    10: "IoT & Embedded Devices Security",
    11: "Red Team/Blue Team & Defense Strategies",
    12: "Cloud Security",
    13: "Wireless Attacks (Advanced)",
    14: "Satellite & Space Infrastructure Security",
    15: "Zero-Day Research & Fuzzing",
    16: "Mobile Security",
    17: "Supply Chain Attacks",
    18: "AI Agents & LLM Exploits",
    19: "Digital Forensics & Incident Response (DFIR)",
    20: "Malware & Threat Analysis",
    21: "Virtualization & Container Security",
    22: "SCADA/ICS Security",
    23: "DevSecOps & CI/CD Security",
    24: "Automotive & Cyber-Physical Systems"
})

print(cluster_category_names)
```

```python
# Si df_counts no está definido, crearlo a partir de df_exploded
if 'df_counts' not in locals():
    df_counts = df_exploded.groupby(['Category_Cluster', 'Impact_Cluster']).size().reset_index(name='count')

# Crear una lista de etiquetas para el eje x en el mismo orden que aparecen en df_counts
x_labels = df_counts['Impact_Cluster'].map(cluster_impact_labels)
y_labels = df_counts['Category_Cluster'].map(cluster_category_names)

plt.figure(figsize=(16, 8))
sns.scatterplot(
    data=df_counts,
    x='Impact_Cluster',
    y='Category_Cluster',
    size='count',
    hue='count',
    sizes=(50, 500),
    palette='viridis',
    legend=False
)

plt.title('Correlación entre Categorías e Impactos')
# Set x-axis labels (Impact Clusters)
plt.xticks(
    ticks=df_counts['Impact_Cluster'].unique(),
    labels=cluster_impact_labels.loc[df_counts['Impact_Cluster'].unique()],
    rotation=90
)
# Set y-axis labels (Category Clusters)
plt.yticks(
    ticks=df_counts['Category_Cluster'].unique(),
    labels=cluster_category_names.loc[df_counts['Category_Cluster'].unique()]
)
plt.tight_layout()
plt.show()
```

Luego de hacer una clusterizacion tanto de las categorias como de los impactos para obtener correlaciones mas generales pero también mas claras, se pudo ver que el impacto más frecuente es la fuga de información ya que guarda una fuerte correlación con varias de las categorías vistas.
Además la ejecución de código, el acceso no autorizado y la filtración de datos se puede ver que son impactos que salen de alguna manera a partir de todas las categorías de los ataques. Entonces se podrían tomar medidas de ciberseguridad para reforzar evitar estos impactos.
