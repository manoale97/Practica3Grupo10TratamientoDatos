```python
# Elimino la columna 15
df = df.drop(columns=['Unnamed: 15'])

# Eliminar las filas con los valores vacios
df = df.dropna()
# Elimina solamente las filas que tengan todos los valores nulos en sus columnas.
# df = df.dropna(how='all')
```

```python
#Tipos de ataques mas comunes
# Guarda en variable attack_counts un conteo de los datos de la columna 'Attack Type' del Dataframe df
attack_counts = df['Attack Type'].value_counts()
print(attack_counts)
```

```python
# Top 10 de ataques mas comunes
import seaborn as sns
import matplotlib.pyplot as plt

# Guarda en variable top10_attacks los datos de mayor valor de attacks_counts
top10_attacks = attack_counts.head(10)
# Asigna al eje 'x' los valores de top10_attacks y al eje 'y' la escala del total de attack_counts.
sns.barplot(x=top10_attacks.index, y=top10_attacks.values, palette='magma')
# Rotacion de texto del eje x.
plt.xticks(rotation=90)
# Titulo de grafico.
plt.title("Top 10 de tipos de ataque mas comunes")
# Dimension del grafico.
plt.figure(figsize=(10, 6))
# Muestra el grafico.
plt.show()
```

Si observamos los tipos de ataques más comunes, el quinto ataque más común es "Malicious Libraries" y el sexto ataque más común es "Malicious Library", lo que se puede entender como un solo dato de tipos de ataques más comunes. Tomarlos por separado podría llevar a errores de análisis, conteos incorrectos y gráficos duplicados, como en el paso anterior.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Reemplaza en la columna Attack Type del dataframe la expresion 'Malicious Library' por 'Malicious Libraries' 
df['Attack Type'] = df['Attack Type'].replace({
    'Malicious Library': 'Malicious Libraries',
})

# Actualiza en variable attack_counts un conteo de los datos de la columna 'Attack Type' del Dataframe df
attack_counts = df['Attack Type'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Guarda en variable top10_attacks los datos de mayor valor de attacks_counts
top10_attacks = attack_counts.head(10)
# Asigna al eje 'x' los valores de top10_attacks y al eje 'y' la escala del total de attack_counts.
sns.barplot(x=top10_attacks.index, y=top10_attacks.values, palette='magma')
# Rotacion de texto del eje x.
plt.xticks(rotation=90)
# Titulo de grafico.
plt.title("Top 10 de tipos de ataque mas comunes")
# Dimension del grafico.
plt.figure(figsize=(10, 6))
# Muestra el grafico.
plt.show()
```

En el análisis anterior se pudo concluir cuál es el tipo de ataque más usual del dataframe analizado; con ese dato se analizará cuál es el objetivo que más aprovechan explotar este tipo de ataque.

```python

# Tambien se puede obtener el tipo de ataque mas comun como texto de la siguiente forma.
most_commom_attack = df['Attack Type'].mode()[0]
print(most_commom_attack)
```

```python
# En el Data Frame se filtran solo las filas que tengan en la columna Attack Type, el ataque mas comun.
df_filtering = df[df['Attack Type'] == most_commom_attack]
df_filtering.head()
```

```python
# Cuenta los dispositivos que mas aprovecha este tipo de ataque.
most_commom_target_type = df_filtering['Target Type'].value_counts()
print("Tipo de dispositivo mas aprovechado por este tipo de ataque:")
print(most_commom_target_type)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Guarda en variable top10_attacks los datos de mayor valor de attacks_counts
top10_most_commom_target_type = most_commom_target_type.head(10)
# Asigna al eje 'x' los valores de top10_attacks y al eje 'y' la escala del total de attack_counts.
sns.barplot(x=top10_most_commom_target_type.index, y=top10_most_commom_target_type.values, palette='magma')
# Rotacion de texto del eje x.
plt.xticks(rotation=90)
# Titulo de grafico.
plt.title("Top 10 de tipos de dispositivos mas aprovechados por ataque Hardware Interface Exploitation")
# Dimension del grafico.
plt.figure(figsize=(10, 6))
# Muestra el grafico.
plt.show()
```
